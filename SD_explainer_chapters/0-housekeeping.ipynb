# Housekeeping and/or troubleshooting

A few notes about these notebooks - these were developed targeting the free tier of the Google Colab cloud platform, specifically the T4 instance (essentially the free GPU). There are also python package resources built into some of these notebooks that need to be installed, but they're a little funny to install on Google Colab. Here's some general housekeeping.


### Some general notes about python notebooks:
If you're new to python and/or python notebooks - firstly, welcome! It's a very accessible coding language, and if you're interested in learning how to code, it's a great option to start with.

Unique to python is a format called "python notebooks", which is what these tutorials are written in. Essentially, python notebooks allow us to run chunks of code, called "cells", one at a time, and then look at any outputs from those cells immediately:



To run a code cell in Google Colab, simply hover over the cell and then click the little play button that pops up. Either that, or you can hit Shift+Enter on your keyboard:



**A BIG NOTE:**
Python (like almost all programming languages) is designed to execute its logic sequentially from top to bottom. Because of this, with these python notebooks, **you must run ALL cells prior to the one you are interested in running.** An easy way to do this by habit is simply hitting Shift+Enter at each cell from the top of the notebook. This automatically scrolls the notebook, and also runs each cell as you go along. If you're having trouble with code not running correctly in your notebook, the first thing you should try is to run everything from top to bottom with no skips. You can do this easily by selecting the "Run all" option from the Runtime dropdown menu:

### GPU Instance:
To make sure the code in these notebooks runs properly, make sure that when you're on Google Colab, you are on a T4 instance. To confirm that you're on one, just check the top right corner of your browser window when you're in google colab. Once you've run a cell, or have clicked "connect", it'll look like this:



** Specifically, look for the "T4" in the top right corner. **

If not, you'll need to switch to the correct runtime. To do so, you'll need to click the drop down arrow in the right hand corner, and click "change runtime type":



From there, you'll get a window pop-up that lets you pick which runtime to use. Under "Hardware Accelerator", pick, "T4 GPU", and click Save:



And then you should be on a T4 instance!

**A NOTE:** on the free tier of Google Colab, you will get a limited amount of compute time on a T4 instance every month (which is nebulously defined by Google dynamically, based on server load). If a T4 instance is unavailable, theoretically these notebooks should still work, but much much slower (BUT THIS IS STILL A WORK IN PROGRESS).

If you are comfortable with python and Jupyter Notebooks (the format of these tutorials), feel free to download these notebooks and run them locally!
